```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train_Green_Energy.csv')
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.info()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df.dtypes
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
```

```python
df.head()
```

```python
df.drop(['datetime','row_id'],axis=1,inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['year','month','hour','day']]
y = df[['energy']]
```

```python
x.head(),y.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
test = pd.read_csv('test_Green_Energy.csv')
```

```python
test.head()
```

```python
test['datetime'] = pd.to_datetime(test['datetime'])
test['year'] = test['datetime'].dt.year
test['month'] = test['datetime'].dt.month
test['day'] = test['datetime'].dt.day
test['hour'] = test['datetime'].dt.hour
```

```python
test.head()
```

```python
x_test = test[['year','month','day','hour']]
y_test = model.predict(x_test)
```

```python
y_test = pd.DataFrame(y_test)
```

```python
Pred_file = pd.concat([test['row_id'],y_test],join= 'outer',axis = 1)
```

```python
Pred_file.columns = ['row_id','energy']
```

```python
Pred_file.head()
```

```python
Pred_file.to_csv('Energy_Predictions_v1.csv',index=False)
```
